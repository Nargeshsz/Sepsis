```python
import pandas as pd
excel_file = "/Users/nargeshosseinzadeh/Downloads/Book1.xlsx"
df = pd.read_excel(excel_file)
```

```python
print(df.head()) 
print(df.info())
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt

## Create a line chart

plt.plot(df['patient_id'], df['T'], marker='o')
plt.xlabel('Patient_id')
plt.ylabel('Temperature')
plt.title('Line Chart')
plt.show()


```

```python
plt.figure(figsize=(20, 10))

# Plotting time series for temperature
plt.subplot(2, 2, 1)
for patient_id, data in df.groupby('patient_id'):
    plt.plot(data['obs_ts'], data['T'], label=f'patient {patient_id}')
plt.xlabel('Date and Time')
plt.ylabel('Temperature')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))

# Plotting time series for HR
plt.subplot(2, 2, 1)
for patient_id, data in df.groupby('patient_id'):
    plt.plot(data['obs_ts'], data['HR'], label=f'patient {patient_id}')
plt.xlabel('Date and Time')
plt.ylabel('Heart Rate')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
# Plotting time series for SBP
plt.subplot(2, 2, 1)
for patient_id, data in df.groupby('patient_id'):
    plt.plot(data['obs_ts'], data['SBP'], label=f'patient {patient_id}')
plt.xlabel('Date and Time')
plt.ylabel('SBP')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(20, 10))
# Plotting time series for DBP
plt.subplot(2, 2, 1)
for patient_id, data in df.groupby('patient_id'):
    plt.plot(data['obs_ts'], data['DBP'], label=f'patient {patient_id}')
plt.xlabel('Date and Time')
plt.ylabel('DBP')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Box plot for temperature
plt.figure(figsize=(8, 6))
df.boxplot(column='T', by='patient_id')
plt.ylabel('Temperature')
plt.title('Temperature Distribution by Patient')
plt.show()
```

```python
# Box plot for Heart Rate
plt.figure(figsize=(8, 6))
df.boxplot(column='HR', by='patient_id')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Distribution by Patient')
plt.show()
```

```python
# Box plot for SBP
plt.figure(figsize=(8, 6))
df.boxplot(column='SBP', by='patient_id')
plt.ylabel('SBP')
plt.title('SBP Distribution by Patient')
plt.show()
```

```python
# Box plot for DBP
plt.figure(figsize=(8, 6))
df.boxplot(column='DBP', by='patient_id')
plt.ylabel('DBP')
plt.title('DBP Distribution by Patient')
plt.show()
```

```python
import seaborn as sns

# Compute the correlation matrix
corr = df[['T', 'HR', 'SBP', 'DBP']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
